```python
df = pca.merge(meta, left_on="#IID", right_on="sample")
```

```python
# Define cohorts for superpopulation (to reduce confounding)
cohorts = {}
for pop in ["EUR", "EAS", "AFR"]:
    cohorts[pop] = df[df["super_pop"] == pop]["#IID"].values
    print(f"Cohort {pop}: {len(cohorts[pop])} individuals")
    # Salva file IID per uso in PLINK2
    pd.DataFrame(cohorts[pop], columns=["IID"]).to_csv(
        f"data/cohort_{pop}.txt", index=False, header=False
    )
```

```python
# Final Report
summary = pd.DataFrame({
    "Cohort": list(cohorts.keys()),
    "N individuals": [len(v) for v in cohorts.values()]
})
summary.to_csv("/mnt/c/Users/alice/Documents/R_scripts/cohort_summary.csv", index=False)
print(summary)
```
